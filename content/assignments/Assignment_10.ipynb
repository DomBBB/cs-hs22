# Task 2

Use the given function `load_iris()` to load the IRIS dataset into a dataframe called `IRIS`.

```python
def load_iris():
    import sklearn.datasets
    iris = sklearn.datasets.load_iris(as_frame=True)
    df = iris["data"].rename(columns={c: c[:-5] for c in iris['data'].columns})
    df['flower'] = iris['target'].apply(lambda f: iris['target_names'][f])
    df = df.set_index('flower')
    return df

IRIS = load_iris()
```
